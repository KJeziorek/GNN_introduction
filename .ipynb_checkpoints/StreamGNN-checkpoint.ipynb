```python
import numpy as np
import os
import torch
import torch_geometric
import cv2

from torch.utils.data.dataset import Dataset
from torch_geometric.data import Data
from torch_geometric.nn import radius_graph, radius
from torch_geometric.transforms import Cartesian 

from typing import List

import time
import networkx as nx
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from tqdm import tqdm
```

```python

class StreamProcess():

  def __init__(self, mode: str = 'FixedEvents', num_events: int = 2500, num_process: int = 1000):

    self.mode = mode # Mode for updating graph: 'FixedEvents' for processing numbers of events specified in num_events and num_process, 'FixedTime' for processing after time specified in time_events and time_process
    
    ### For FixedEvents ###
    self.num_events = num_events # Numbers of events in graph after init
    self.num_process = num_process # Numbers of events after graph will be processed

    self.x = torch.tensor([], dtype=torch.int)
    self.pos = torch.tensor([], dtype=torch.float)
    self.edge_index = torch.tensor([[],[]], dtype=torch.long)
    self.edge_attr = torch.tensor([], dtype=torch.float)

    self.r = 3
    self.iter = 0

  def add_event(self, x: int, pos: List[int]):
    # Count added events
    self.iter += 1

    # Convert arrays to tensors 
    x_new = torch.tensor([x], dtype=torch.int)
    pos[2] *= 0.5e+3 # normalize time
    pos_new = torch.tensor([pos], dtype=torch.float)

    if self.x.numel() > self.num_events:
      # Delete oldest node with edges from graph
      self.x = self.x[:-1]
      self.pos = self.pos[:-1]

      edge_index = self.edge_index.numpy()
      edge_attr = self.edge_attr.numpy()

      to_delete = (edge_index==self.num_events).nonzero()[1]

      self.edge_index = torch.from_numpy(np.delete(edge_index, to_delete, 1))
      self.edge_attr = torch.from_numpy(np.delete(edge_attr, to_delete, 0))


    if self.x.numel() == 0:
      # Add first node to graph
      self.x = torch.cat((x_new, self.x), 0)  
      self.pos = torch.cat((pos_new, self.pos), 0)
    else:
      self.edge_index += 1

      # Find neighbours for new_event based on radius
      edge_index_new = radius(pos_new, self.pos, r=self.r)

      # Increment all index for edges that are currently in graph
      edge_index_new[0] += 1

      # Swap index to make undirected graph
      swaped_edge_indeg_new = torch.index_select(edge_index_new, 0, torch.LongTensor([1, 0]))
      
      # Merge both tensor
      cat_new_edge = torch.cat((edge_index_new, swaped_edge_indeg_new), -1)
      
      # Add x, pos and edge_index to tensors
      self.x = torch.cat((x_new, self.x), 0)  
      self.pos = torch.cat((pos_new, self.pos), 0)
      self.edge_index = torch.cat((cat_new_edge, self.edge_index), -1)

      # Based on "radius" function from PyG
      (row, col) = cat_new_edge
      cart = self.pos[row] - self.pos[col]

      if cart.numel() > 0:
          cart = cart / (2 * self.r) + 0.5

      # Add edge attribute to tensor
      self.edge_attr = torch.cat((cart, self.edge_attr), 0)

  def result(self):
    return Data(x=self.x, pos=self.pos, edge_index=self.edge_index, edge_attr=self.edge_attr) 
```

```python
process = StreamProcess()

events = np.loadtxt('events.txt')
xv, posv = events[:, -1:], events[:, :3]

t = time.time()
for x, pos in zip(xv, posv):
    process.add_event(x=x, pos=pos)
print("Process time = ", time.time()-t, "s")

data = process.result()
print(data)
```

```python
fig, ax = plt.subplots(1, 1)

pos_x = data.pos[:, 0]
pos_y = data.pos[:, 1]

vec = np.zeros((72,81))
for x, y, p in zip(pos_x, pos_y, data.x):
  vec[int(y), int(80-x)] = 255

plt.imshow(vec)
plt.show()


if (edge_index := getattr(data, "edge_index")) is not None:
    for edge in tqdm(edge_index.T):
        pos_edge = data.pos[[edge[0], edge[1]], :]
        ax.plot(pos_edge[:, 0], pos_edge[:, 1], "k-", linewidth=0.1)
ax.scatter(pos_x, pos_y, s=2)

fig.tight_layout()
plt.show()


# Create the 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Plot the nodes - alpha is scaled by "depth" automatically


x, y, t = data.pos.numpy().T
ax.scatter(x, t, y, s=5, ec="w", c=data.x, cmap='bwr')

edge_xyz = np.array([(data.pos[u].numpy(), data.pos[v].numpy()) for u, v in torch.transpose(data.edge_index, 0, 1)])

# Plot the edges
for vizedge in edge_xyz:
    new_vizedge = vizedge.copy()
    new_vizedge[:,0], new_vizedge[:,1], new_vizedge[:,2] = vizedge[:,0], vizedge[:,2], vizedge[:,1]
    ax.plot(*new_vizedge.T, "k-", linewidth=0.1)


def _format_axes(ax):
    """Visualization options for the 3D axes."""
    # Turn gridlines off
    ax.grid(False)
    # Suppress tick labels
    for dim in (ax.xaxis, ax.yaxis, ax.zaxis):
        dim.set_ticks([])
    # Set axes labels
    ax.set_xlabel("x")
    ax.set_ylabel("t")
    ax.set_zlabel("y")


_format_axes(ax)
fig.tight_layout()
plt.show()
```
